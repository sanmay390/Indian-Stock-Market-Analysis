```python
#!pip install fbprophet
from fbprophet import Prophet
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from nsepy import get_history
from datetime import date
data = get_history(symbol='IRCTC',
                        start=date(2020,5,1), 
                        end=date(2021,2,20))
```

```python
#type(sbi)
df=pd.DataFrame(data['Close'])
df.plot()
plt.show()
```
